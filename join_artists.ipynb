```python
import pandas as pd
import string
```

```python
data = pd.read_csv("dataset/data.csv")
```

```python
genres = pd.read_csv("artist_genre.csv")
```

```python
exploded = data.assign(artists = data.artists.str.split(",")).explode('artists')
# exploded.join(genres, on = "artists")
exploded["artists"] = exploded["artists"].apply(lambda x: str(x).translate(str.maketrans('', '', string.punctuation)))
merged = exploded.merge(genres, left_on = exploded["artists"].astype(str), right_on = genres["artists"].astype(str), how = "inner")
merged = merged.drop(['key_0', 'artists_y', 'artists_x', 'explicit'], axis=1)
```

```python
merged = merged[merged.new_genres.astype(str) != '[\'other\']']
merged = merged.drop_duplicates(subset = ['id'])
merged.sample(5)
```

```python
exp = merged.assign(new_genres = merged.new_genres.str.split(",")).explode('new_genres')
# exploded.join(genres, on = "artists")
exp["new_genres"] = exp["new_genres"].apply(lambda x: str(x).translate(str.maketrans('', '', string.punctuation)))
```
